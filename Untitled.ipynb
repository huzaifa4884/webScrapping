```python
# imports
import requests
import bs4
import csv
```

```python
res = requests.get("https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops")
```

```python
soup = bs4.BeautifulSoup(res.text , 'lxml')
```

```python
#extracting thumbnails 
thumbs = soup.find_all('div' , class_='thumbnail')
```

```python
print(thumbs[0].find('h4').text)
print(thumbs[0].find('a').text)
print(thumbs[0].find('p').text)
print(thumbs[0].find('div' , class_ = 'ratings').text.replace("\n" , ''))
```

```python
price = []
item_name = []
description = []
reviews = []

for thumb in thumbs:
    price.append(thumb.h4.text)
    item_name.append(thumb.a.text)
    description.append(thumb.p.text)
    reviews.append(thumb.find('div' , class_= 'ratings').text.replace("\n",""))
```

```python
for i , j , k ,l in zip(item_name , price , description , reviews):
    print(f'''
     Item_Name = {i}
     Price = {j}
     Description = {k}
     Reviews = {l}
    ''')
```

```python
with open('data.csv' , 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['items' , 'price' , 'description' , 'reviews'])
    for i in range(len(item_name)):
        writer.writerow([item_name[i] , price[i] , description[i] , reviews[i]])
```
